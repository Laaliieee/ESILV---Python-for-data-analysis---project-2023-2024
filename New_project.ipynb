Consumption of illegal drugs by men and womens

```python
def create_gender_consumption(ax, male_users, female_users, drug_columns):

    drug_freq_df = pd.DataFrame(index=drug_columns)
    drug_freq_df['Male'] = male_users[drug_columns].apply(lambda x: (x != 'Never Used').sum())
    drug_freq_df['Female'] = female_users[drug_columns].apply(lambda x: (x != 'Never Used').sum())

    
    sns.set(style="whitegrid")
    plot = drug_freq_df.plot(kind='bar', ax=ax, color=['c', 'r'])
    
    ax.set_xticklabels(drug_freq_df.index, rotation=45, ha='right')
    ax.set_xlabel('Drug')
    ax.set_ylabel('Consumption')
    ax.set_title('Consumption of different drug by genre')
    ax.legend(title='Genre', loc='upper right')

fig, ax = plt.subplots(figsize=(12, 8))
create_gender_consumption(ax, male, female, illegal_drug)
plt.show()
```

## Decision Tree Classifier

We prepare the dataset in two different sets 

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test=train_test_split(X, Y, test_size= 0.2, random_state = 5)
```

```python
train_data = pd.concat([pd.DataFrame(X_train, columns=legal_drug), pd.Series(Y_train, name='gender')], axis=1)

sns.pairplot(train_data, hue='gender', palette='viridis')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
model = DecisionTreeClassifier(random_state=5)
model.fit(X_train, Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
predictions=model.predict(X_test)
```

```python
def gender (model,alcohol,caff,choc,legalh,nicotine):
    x=np.array([alcohol,caff,choc,legalh,nicotine]).reshape(1,5)
    print(model.predict(x))
```

```python
gender(model, 1, 1, 3,0,0) #Test for the first line of the dataset, here we have a good prediction
```

```python
def gender_proba (model,alcohol,caff,choc,legalh,nicotine):
    x=np.array([alcohol,caff,choc,legalh,nicotine]).reshape(1,5)
    print(model.predict(x))
    print(model.predict_proba(x))
```

```python
gender_proba(model, 1, 1, 3,0,0) #Test for the first line of the dataset, here we have a good probability 
```

The responce given by the algo suggest that at 80% we can say that the person is a male which is the good answer

Let us see the score of the algorithm

```python
from sklearn.model_selection import cross_val_score

cross_val_score(DecisionTreeClassifier(),X_train,Y_train, cv=5, scoring='accuracy')
```

The model don't have good accuracy score, let us see if we can increase the performances of the model with the gread search

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

param_grid = {
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, Y_train)

# Meilleurs paramètres
print(grid_search.best_params_)
```

```python
best_dt_classifier = grid_search.best_estimator_
best_dt_classifier.fit(X_train,Y_train)
# Make predictions on the test set using the best model
predictions = best_dt_classifier.predict(X_test)

# Evaluate the performance of the best classifier
accuracy = accuracy_score(Y_test, predictions)
classification_rep = classification_report(Y_test, predictions)
```
